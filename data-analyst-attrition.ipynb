### Uji inferensial

Uji inferensial digunakan untuk mengetahui apakah ada **perbedaan signifikan** antara karyawan yang resign dan tidak resign berdasarkan karakteristik personal (usia, gender, marital status, education). Jenis uji yang digunakan tergantung pada tipe data setiap variabel.

Sebelumnya dilakukan Saphiro test untuk kolom numerik (age) untuk mengetahui apakah datanya terdistribusi normal

```python
from scipy.stats import shapiro
import pandas as pd

# Load data
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")

# Ambil data usia berdasarkan status Attrition
resign_age = df[df['Attrition'] == 'Yes']['Age']
stay_age = df[df['Attrition'] == 'No']['Age']

# Lakukan uji Shapiro-Wilk
shapiro_resign = shapiro(resign_age)
shapiro_stay = shapiro(stay_age)

# Tampilkan hasil
print("=== Shapiro-Wilk Test: Age ===")
print(f"Resign - W: {shapiro_resign.statistic:.4f}, p-value: {shapiro_resign.pvalue:.4f}")
print(f"Stay   - W: {shapiro_stay.statistic:.4f}, p-value: {shapiro_stay.pvalue:.4f}")
```

✅ Kesimpulan:
- Karena p-value < 0.05 untuk kedua kelompok, maka kita menolak H0.
- Artinya, distribusi usia di kedua kelompok tidak normal secara statistik.

### Uji Inferensial yang Dilakukan

#### 1. Usia (Age)
- Jenis uji : Mann-Whitney U Test
- Alasan : Age adalah variabel numerik, uji ini digunakan untuk membandingkan rata-rata usia antara dua kelompok (resign dan tidak resign) dan distribusi data tidak normal.

#### 2. Gender
- Jenis uji: **Chi-Square Test of Independence**
- Alasan: Gender merupakan variabel kategorikal, dan uji ini digunakan untuk mengetahui apakah **gender berpengaruh secara signifikan terhadap kemungkinan resign**. Dengan kata lain, apakah laki-laki dan perempuan memiliki kecenderungan resign yang berbeda secara statistik.

#### 3. Marital Status
- Jenis uji: **Chi-Square Test of Independence**
- Alasan: Sama seperti gender, marital status adalah variabel kategorikal. Uji chi-square digunakan untuk mengecek apakah **status pernikahan berkorelasi signifikan** dengan attrition.

#### 4. Education
- Jenis uji: **Chi-Square Test of Independence**
- Alasan: Education pada dataset berupa skor ordinal, namun dianggap kategorikal dalam analisis ini. Tujuannya untuk melihat apakah **jenjang pendidikan memengaruhi** keputusan resign secara signifikan.

```python
# =============================
#  Uji Statistik Inferensial
# =============================

from scipy.stats import mannwhitneyu, chi2_contingency
import pandas as pd

# Load data
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")

# Mann-Whitney U Test: Age (karena data usia tidak terdistribusi normal)
resign_age = df[df['Attrition'] == 'Yes']['Age']
stay_age = df[df['Attrition'] == 'No']['Age']
u_stat_age, p_value_age = mannwhitneyu(resign_age, stay_age, alternative='two-sided')

# Chi-Square: Gender
chi2_gender, p_gender, _, _ = chi2_contingency(pd.crosstab(df['Gender'], df['Attrition']))

# Chi-Square: Marital Status
chi2_marital, p_marital, _, _ = chi2_contingency(pd.crosstab(df['MaritalStatus'], df['Attrition']))

# Chi-Square: Education
chi2_edu, p_edu, _, _ = chi2_contingency(pd.crosstab(df['Education'], df['Attrition']))

# =====================
# Ringkasan Hasil
# =====================
print("=== Mann-Whitney U Test: Age ===")
print(f"U Statistic: {u_stat_age:.2f}, P-Value: {p_value_age:.4f}")

print("\n=== Chi-Square Test: Gender ===")
print(f"Chi2: {chi2_gender:.2f}, P-Value: {p_gender:.4f}")

print("\n=== Chi-Square Test: Marital Status ===")
print(f"Chi2: {chi2_marital:.2f}, P-Value: {p_marital:.4f}")

print("\n=== Chi-Square Test: Education ===")
print(f"Chi2: {chi2_edu:.2f}, P-Value: {p_edu:.4f}")
```

#### 🧪 Hasil dan Analsis :
| No | Variabel        | Jenis Uji     | p-value     | Signifikan? | Kesimpulan                                                                 |
|----|------------------|----------------|-------------|--------------|------------------------------------------------------------------------------|
| 1  | Age              | Mann-Whitney        | < 0.05      |  Ya        | Ada perbedaan usia yang signifikan antara yang resign dan yang tidak         |
| 2  | Gender           | Chi-Square     | > 0.05      |  Tidak     | Gender tidak berpengaruh signifikan terhadap attrition           |
| 3  | MaritalStatus    | Chi-Square     | < 0.05      |  Ya        | Status pernikahan berpengaruh signifikan terhadap keputusan resign          |
| 4  | Education        | Chi-Square     | > 0.05      |  Tidak     | Tingkat pendidikan tidak berpengaruh signifikan terhadap attrition          |

### 🧠 Kesimpulan:
Faktor demografi yang **paling berpengaruh terhadap keputusan resign** adalah:
- **Usia** → karyawan usia muda cenderung lebih banyak resign
- **Status pernikahan** → karyawan yang masih Single lebih banyak meninggalkan perusahaan

Sedangkan **jenis kelamin dan tingkat pendidikan** tidak menunjukkan pengaruh signifikan. Maka dari itu, strategi retensi karyawan dapat difokuskan pada segmen karyawan muda dan belum menikah yang berisiko tinggi untuk melakukan resign.

Untuk menjawab pertanyaan ini, dilakukan perbandingan antara karyawan usia 26–30 yang *resign* dan yang *tidak resign*, berdasarkan faktor-faktor berikut:

### Uji Inferensial

Analisis ini bertujuan untuk mengetahui apakah terdapat perbedaan karakteristik yang signifikan antara karyawan usia 26–30 tahun yang resign dan yang bertahan di perusahaan terhadap beberapa aspek.

Sebelumnya, dilakukan uji distribusi  normalitas terhadap kolom numerik (Monthly Income dan Distance from Home) dengan uji Saphiro-Wilk

```python
from scipy.stats import shapiro

# Misalnya df adalah DataFrame utama
# Filter karyawan usia muda, misalnya < 30 tahun
young_employees = df[df['Age'] < 30]

# Bagi menjadi dua kelompok: resign dan stay
resign_young = young_employees[young_employees['Attrition'] == 'Yes']
stay_young = young_employees[young_employees['Attrition'] == 'No']

# Uji normalitas Monthly Income
shapiro_income_resign = shapiro(resign_young['MonthlyIncome'])
shapiro_income_stay = shapiro(stay_young['MonthlyIncome'])

# Uji normalitas Distance From Home
shapiro_distance_resign = shapiro(resign_young['DistanceFromHome'])
shapiro_distance_stay = shapiro(stay_young['DistanceFromHome'])

# Cetak hasil uji Shapiro
print("=== Shapiro–Wilk Test ===")
print(f"MonthlyIncome (Resign): W = {shapiro_income_resign.statistic:.4f}, p = {shapiro_income_resign.pvalue:.4f}")
print(f"MonthlyIncome (Stay):   W = {shapiro_income_stay.statistic:.4f}, p = {shapiro_income_stay.pvalue:.4f}")
print(f"Distance (Resign):      W = {shapiro_distance_resign.statistic:.4f}, p = {shapiro_distance_resign.pvalue:.4f}")
print(f"Distance (Stay):        W = {shapiro_distance_stay.statistic:.4f}, p = {shapiro_distance_stay.pvalue:.4f}")
```

- Karena semua p-value < 0.05, maka seluruh variabel tidak berdistribusi normal.
- Uji yang digunakan untuk kolom numerik (Monthly Income dan Distance from Home) adalah Mann-Whitney U Test

### Uji Inferensial yang dilakukan :

#### 1. Mann-Whitney U Test
- **Variabel yang diuji:** MonthlyIncome, DistanceFromHome
- **Tujuan:** Menguji apakah terdapat perbedaan signifikan antara pendapatan dan jarak rumah ke kantor antara karyawan yang resign dan tidak resign.
- **Alasan:** Kedua variabel bersifat numerik dan dibandingkan antar dua kelompok, distribusi data tidak normal, digunakan untuk mengetahui apakah perbedaan rata-rata tersebut signifikan secara statistik.

#### 2. Chi-Square Test of Independence
- **Variabel yang diuji:** OverTime, BusinessTravel
- **Tujuan:** Menguji apakah ada hubungan signifikan antara status lembur atau frekuensi perjalanan dinas dengan keputusan resign.
- **Alasan:** Kedua variabel bersifat kategorikal; chi-square digunakan untuk melihat apakah distribusi attrition berbeda signifikan antar kategori.

#### 3. Kruskal-Wallis Test
- **Variabel yang diuji:** JobSatisfaction, WorkLifeBalance
- **Tujuan:** Menguji apakah tingkat kepuasan kerja dan work-life balance berpengaruh terhadap attrition secara signifikan.
- **Alasan:** Kedua variabel bersifat ordinal (skala 1–4) dan tidak diasumsikan berdistribusi normal; Kruskal-Wallis digunakan sebagai alternatif non-parametrik dari ANOVA.

```python
from scipy.stats import mannwhitneyu, chi2_contingency, kruskal
import pandas as pd

# Load data
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")

# Filter usia 26–30
group_young = df[df['Age'].between(26, 30)]
resign_young = group_young[group_young['Attrition'] == 'Yes']
stay_young = group_young[group_young['Attrition'] == 'No']

# 1. Mann-Whitney U Test untuk numerik (karena tidak normal)
u_income, p_income = mannwhitneyu(resign_young['MonthlyIncome'], stay_young['MonthlyIncome'], alternative='two-sided')
u_distance, p_distance = mannwhitneyu(resign_young['DistanceFromHome'], stay_young['DistanceFromHome'], alternative='two-sided')

# 2. Chi-Square Test
crosstab_ot = pd.crosstab(group_young['Attrition'], group_young['OverTime'])
crosstab_bt = pd.crosstab(group_young['Attrition'], group_young['BusinessTravel'])

if crosstab_ot.shape[0] > 1 and crosstab_ot.shape[1] > 1:
    chi2_ot, p_ot, _, _ = chi2_contingency(crosstab_ot)
else:
    chi2_ot, p_ot = None, None
    print("❗ Crosstab OverTime tidak valid untuk Chi-Square.")

if crosstab_bt.shape[0] > 1 and crosstab_bt.shape[1] > 1:
    chi2_bt, p_bt, _, _ = chi2_contingency(crosstab_bt)
else:
    chi2_bt, p_bt = None, None
    print("❗ Crosstab BusinessTravel tidak valid untuk Chi-Square.")

# 3. Kruskal-Wallis Test untuk ordinal
k_js = kruskal(resign_young['JobSatisfaction'], stay_young['JobSatisfaction'])
k_wlb = kruskal(resign_young['WorkLifeBalance'], stay_young['WorkLifeBalance'])

# =====================
# Ringkasan Hasil
# =====================
print("\n=== Mann-Whitney U Test ===")
print(f"MonthlyIncome: U = {u_income:.4f}, p = {p_income:.4f}")
print(f"DistanceFromHome: U = {u_distance:.4f}, p = {p_distance:.4f}")

print("\n=== Chi-Square ===")
if p_ot is not None:
    print(f"OverTime: chi2 = {chi2_ot:.4f}, p = {p_ot:.4f}")
else:
    print("OverTime: Tidak bisa diuji (crosstab tidak valid)")

if p_bt is not None:
    print(f"BusinessTravel: chi2 = {chi2_bt:.4f}, p = {p_bt:.4f}")
else:
    print("BusinessTravel: Tidak bisa diuji (crosstab tidak valid)")

print("\n=== Kruskal-Wallis ===")
print(f"JobSatisfaction: H = {k_js.statistic:.4f}, p = {k_js.pvalue:.4f}")
print(f"WorkLifeBalance: H = {k_wlb.statistic:.4f}, p = {k_wlb.pvalue:.4f}")
```

## Uji Inferensial
- Berdasarkan analisis sebelumnya, departemen Sales diketahui memiliki attrition rate tertinggi secara signifikan.
- Oleh karena itu, dilakukan uji inferensial lanjutan secara khusus terhadap karyawan di departemen Sales saja.

Sebelumnya, dilakukan uji distribusi  normalitas terhadap kolom numerik (Monthly Income dan Distance from Home) dengan uji Saphiro-Wilk

```python
from scipy.stats import shapiro
import pandas as pd

# Load dataset
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")

# Filter data Sales
sales_df = df[df['Department'] == 'Sales']
resign_sales = sales_df[sales_df['Attrition'] == 'Yes']
stay_sales = sales_df[sales_df['Attrition'] == 'No']

# Shapiro-Wilk Test
print("=== Shapiro-Wilk Test ===")

# MonthlyIncome
w_income_resign, p_income_resign = shapiro(resign_sales['MonthlyIncome'])
print(f"MonthlyIncome - Resign: W = {w_income_resign:.4f}, p = {p_income_resign:.10f}")

w_income_stay, p_income_stay = shapiro(stay_sales['MonthlyIncome'])
print(f"MonthlyIncome - Stay:   W = {w_income_stay:.4f}, p = {p_income_stay:.10f}")

# DistanceFromHome
w_distance_resign, p_distance_resign = shapiro(resign_sales['DistanceFromHome'])
print(f"DistanceFromHome - Resign: W = {w_distance_resign:.4f}, p = {p_distance_resign:.10f}")

w_distance_stay, p_distance_stay = shapiro(stay_sales['DistanceFromHome'])
print(f"DistanceFromHome - Stay:   W = {w_distance_stay:.4f}, p = {p_distance_stay:.10f}")
```

- Seluruh data numerik pada departemen Sales baik untuk yang resign maupun tidak resign tidak berdistribusi normal
- Uji Mann-Whitney U Test untuk membandingkan karakteristiknya

## Uji Inferensial yang dilakukan :

#### 1. Mann-Whitney U Test
- **Variabel yang diuji:** MonthlyIncome, DistanceFromHome
- **Tujuan:** Menguji apakah terdapat perbedaan signifikan antara pendapatan dan jarak rumah ke kantor antara karyawan yang resign dan tidak resign.
- **Alasan:** Kedua variabel bersifat numerik dan dibandingkan antar dua kelompok, distribusi data tidak normal, digunakan untuk mengetahui apakah perbedaan rata-rata tersebut signifikan secara statistik.

#### 2. Chi-Square Test of Independence
- **Variabel yang diuji:** OverTime, BusinessTravel
- **Tujuan:** Menguji apakah ada hubungan signifikan antara status lembur atau frekuensi perjalanan dinas dengan keputusan resign.
- **Alasan:** Kedua variabel bersifat kategorikal; chi-square digunakan untuk melihat apakah distribusi attrition berbeda signifikan antar kategori.

#### 3. Kruskal-Wallis Test
- **Variabel yang diuji:** JobSatisfaction, WorkLifeBalance
- **Tujuan:** Menguji apakah tingkat kepuasan kerja dan work-life balance berpengaruh terhadap attrition secara signifikan.
- **Alasan:** Kedua variabel bersifat ordinal (skala 1–4) dan tidak diasumsikan berdistribusi normal; Kruskal-Wallis digunakan sebagai alternatif non-parametrik dari ANOVA.

```python
from scipy.stats import mannwhitneyu, chi2_contingency, kruskal

# Filter hanya data di departemen Sales
sales_df = df[df['Department'] == 'Sales']

# Pisahkan berdasarkan Attrition
resign = sales_df[sales_df['Attrition'] == 'Yes']
stay = sales_df[sales_df['Attrition'] == 'No']

# --- Mann-Whitney U Test (numerik) ---
u_income, p_income = mannwhitneyu(resign['MonthlyIncome'], stay['MonthlyIncome'], alternative='two-sided')
u_distance, p_distance = mannwhitneyu(resign['DistanceFromHome'], stay['DistanceFromHome'], alternative='two-sided')

# --- Chi-Square (kategorikal) ---
chi2_overtime, p_overtime, _, _ = chi2_contingency(pd.crosstab(sales_df['OverTime'], sales_df['Attrition']))
chi2_travel, p_travel, _, _ = chi2_contingency(pd.crosstab(sales_df['BusinessTravel'], sales_df['Attrition']))

# --- Kruskal-Wallis (ordinal) ---
kw_stat_satisfaction, p_satisfaction = kruskal(resign['JobSatisfaction'], stay['JobSatisfaction'])
kw_stat_wlb, p_wlb = kruskal(resign['WorkLifeBalance'], stay['WorkLifeBalance'])

# --- Tampilkan hasil ---
print("=== Mann-Whitney U Test ===")
print(f"MonthlyIncome p-value: {p_income:.6f}")
print(f"DistanceFromHome p-value: {p_distance:.6f}\n")

print("=== Chi-Square ===")
print(f"OverTime p-value: {p_overtime:.6f}")
print(f"BusinessTravel p-value: {p_travel:.6f}\n")

print("=== Kruskal-Wallis ===")
print(f"JobSatisfaction p-value: {p_satisfaction:.6f}")
print(f"WorkLifeBalance p-value: {p_wlb:.6f}")
```

## 5. Bagaimana attrition rate berdasarkan durasi kerja karyawan dalam berbagai aspek?

### 🎯 Tujuan:
Mengetahui apakah terdapat pola attrition berdasarkan **lama pengalaman kerja** karyawan, baik secara keseluruhan maupun dalam peran saat ini, sejak terakhir promosi, dan dengan manajer saat ini.

**Variabel yang dianalisis**:
- YearsAtCompany
- YearsInCurrentRole
- YearsSinceLastPromotion

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load data dan siapkan binary flag
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df['AttritionFlag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Buat kategori durasi kerja
df['YearsAtCompanyGroup'] = pd.cut(df['YearsAtCompany'], bins=[0, 3, 6, 10, 40], labels=['0-3', '4-6', '7-10', '11+'])
df['YearsInCurrentRoleGroup'] = pd.cut(df['YearsInCurrentRole'], bins=[0, 2, 4, 7, 18], labels=['0-2', '3-4', '5-7', '8+'])
df['YearsSinceLastPromotionGroup'] = pd.cut(df['YearsSinceLastPromotion'], bins=[-1, 0, 2, 5, 15], labels=['0', '1-2', '3-5', '6+'])
df['YearsWithCurrManagerGroup'] = pd.cut(df['YearsWithCurrManager'], bins=[0, 2, 4, 7, 20], labels=['0-2', '3-4', '5-7', '8+'])

# Plot bar chart
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

group_vars = {
    'YearsAtCompanyGroup': 'Years at Company',
    'YearsInCurrentRoleGroup': 'Years in Current Role',
    'YearsSinceLastPromotionGroup': 'Years Since Last Promotion',
    'YearsWithCurrManagerGroup': 'Years with Current Manager'
}

for i, (col, title) in enumerate(group_vars.items()):
    grouped = df.groupby(col, observed=False)['AttritionFlag'].mean() * 100
    sns.barplot(x=grouped.index, y=grouped.values, ax=axes[i])
    axes[i].set_title(f'Attrition Rate by {title}')
    axes[i].set_ylabel('Attrition Rate (%)')
    axes[i].set_xlabel(title)

plt.tight_layout()
plt.show()
```

  - Semakin lama masa kerja, cenderung attrition rate-nya makin rendah.

## Uji Inferensial

- Uji inferensial dilakukan untuk mengetahui apakah variabel durasi kerja memiliki hubungan signifikan dengan keputusan resign (attrition).
- Karena semua variabel durasi bersifat numerik dan tidak berdistribusi normal, maka digunakan uji korelasi non-parametrik.

#### Uji yang Digunakan: Spearman Rank Correlation
- **Variabel yang diuji:**
  - YearsAtCompany
  - YearsInCurrentRole
  - YearsWithCurrManager
  - YearsSinceLastPromotion
- **Tujuan:** Untuk mengukur kekuatan dan arah hubungan antara masing-masing variabel durasi kerja dengan kemungkinan resign.
- **Alasan:**
  - Variabel durasi bersifat numerik.
  - Variabel attrition dikonversi menjadi binary (Yes = 1, No = 0).
  - Spearman digunakan karena tidak mengasumsikan distribusi normal, dan mampu mendeteksi hubungan monotonik antar variabel.

```python
from scipy.stats import spearmanr

# Tentukan kolom-kolom durasi kerja
duration_cols = [
    'YearsAtCompany',
    'YearsInCurrentRole',
    'YearsSinceLastPromotion',
    'YearsWithCurrManager'
]

# Uji korelasi Spearman antara tiap kolom durasi dan AttritionFlag
for col in duration_cols:
    coef, pval = spearmanr(df[col], df['AttritionFlag'])
    print(f"{col:<25} → Spearman rho: {coef:.4f} | p-value: {pval:.4f}")
```

- Tawarkan kesempatan kerja yang lebih dinamis atau variatif kepada karyawan dengan peran yang cenderung monoton atau jarang bepergian. Proyek lintas tim dan penugasan lapangan bisa menjadi alternatif.
- Perkuat hubungan antara karyawan dan manajer melalui komunikasi rutin dan coaching berkala. Hal ini penting terutama pada fase awal hubungan kerja.


## 🚀 Dampak Bisnis dari Implementasi Strategi Retensi Karyawan

1. **📉 Penurunan Biaya Turnover**
   - Mengurangi biaya rekrutmen, pelatihan, dan adaptasi karyawan baru.
   - Menghemat 33%–200% dari gaji tahunan setiap karyawan yang tidak jadi resign (SHRM estimate).
   - Biaya tidak langsung seperti hilangnya pengetahuan organisasi juga ikut ditekan.

2. **📈 Produktivitas Karyawan Lebih Konsisten**
   - Karyawan yang lebih engaged, tidak kelelahan karena lembur berlebih, dan merasa dihargai, cenderung menunjukkan performa lebih tinggi.
   - Divisi rawan seperti Sales bisa meningkatkan pencapaian target karena tim lebih stabil.

3. **🤝 Loyalitas dan Kepuasan Karyawan Meningkat**
   - Program retensi dan perhatian terhadap WLB serta career development memperkuat employee experience.
   - Loyalitas meningkat → menciptakan budaya kerja positif → jadi magnet bagi talenta baru (employer branding membaik).

4. **📊 Performa Divisi Kritis Lebih Stabil**
   - Departemen dengan turnover tinggi seperti Sales dan HR jadi lebih solid dan minim distraksi.
   - Pelanggan dan mitra bisnis akan melihat konsistensi dalam pelayanan dan hubungan.

5. **💡 Lebih Banyak Insight dari Internal Talent**
   - Karyawan yang stay lebih lama bisa dikembangkan jadi pemimpin masa depan (succession planning lebih kuat).
   - Perusahaan tidak bergantung pada perekrutan eksternal untuk peran strategis.

6. **🏆 Keunggulan Kompetitif di Industri**
   - Retensi yang tinggi = karyawan lebih berpengalaman dan adaptif terhadap strategi perusahaan.
   - Perusahaan lebih siap menjalankan proyek jangka panjang dengan tim yang terjaga.
